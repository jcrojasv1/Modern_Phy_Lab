```python
import pandas as pd
import numpy as np
import sympy as sym
from scipy import constants
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def GetRegression(x,y,e):
    p, pcov = np.polyfit(x,y,1, w=1./e, cov = "unscaled")
    m,b = p
    X = np.linspace(min(x),max(x),100)
    Y = m*X + b
    return X,Y,m,b, pcov

def GetR2(m, b, x, y):
    # Cálculo de la "bondad" del ajuste con base en el parámetro R^2:
    mean_y = np.sum(y)/len(y)
    reg_y = m*x + b
    SS_res = np.sum((y - reg_y)**2)
    SS_tot = np.sum((y - mean_y)**2)
    Rp2 = 1 - SS_res/SS_tot
    return Rp2
```

```python
# Información para el Hierro:

Fe_Vol = 2.1*2.075*2.1 # Valor principal 
Fe_Vol_Inc = (0.005/2.1 + 0.005/2.075 + 0.005/2.1)*Fe_Vol
Fe_A, Fe_B = 0.594, 0.509

Fe_Vol, Fe_Vol_Inc
```

```python
Fe_Vol_Exp = (Fe_A - Fe_B)/(997*9.81) * 1e6 # Para pasar de m^3 a cm^3
Fe_Vol_Exp_Inc = np.sqrt(2*(0.005)**2)
Fe_Vol_Exp, Fe_Vol_Exp_Inc
```

```python
# Información para el Latón:

La_Vol = 2.1*2.1*1.82 # Valor principal 
La_Vol_Inc = (0.005/2.1 + 0.005/1.82 + 0.005/2.1)*La_Vol
La_A, La_B = 0.593, 0.525

La_Vol, La_Vol_Inc
```

```python
La_Vol_Exp = (La_A - La_B)/(997*9.81) * 1e6 # Para pasar de m^3 a cm^3
La_Vol_Exp_Inc = np.sqrt(2*(0.005)**2)
La_Vol_Exp, La_Vol_Exp_Inc
```

```python
# Información para el Aluminio:
F_N = np.array([233.8 - 229.8, 235.4 - 229.8, 238.9 - 229.8, 244.5 - 229.8, 246.7 - 229.8, 270 - 229.8])*1e-3*9.81
V = np.array([5.62 - 4.35, 5.62 - 3.77, 5.62 - 3, 5.62 - 1.6, 5.62 - 1.16, 5.62])*4.41*10
T = np.array([0.555, 0.548, 0.515, 0.456, 0.437, 0.353])
Al_Vol = 5.62*2.1*2.1
Al_Vol_Inc = (0.005/2.1 + 0.005/5.62 + 0.005/2.1)*Al_Vol
```

```python
plt.figure(figsize = (8, 5))
e1 = np.array([0.001]*6)
_V, _T, m1, b1, pcov1 = GetRegression(V, T, e1)
R2_Al1 = GetR2(m1, b1, V, T)
plt.plot(_V*1e-1, _T, color = "#6D1305")
plt.scatter(V*1e-1, T, color = "#C70039")
perr1 = np.sqrt(np.diag(pcov1))
plt.grid()
plt.xlabel(r"Volumen Sumergido [cm$^{3}$]")
plt.ylabel(r"Tensión [N]")
plt.text(17.5, 0.50, r"R$^{2}$ = 0.929", color = "#C70039", fontsize = 13)
plt.savefig("Tensión.pdf")
m1, perr1[0], R2_Al1
```

```python
plt.figure(figsize = (8, 5))
e2 = np.array([0.01]*5)
_V_, _F_N, m2, b2, pcov2 = GetRegression(V[0:5], F_N[0:5], e2)
R2_Al2 = GetR2(m2, b2, V[0:5], F_N[0:5])
plt.plot(_V_*1e-1, _F_N, color = "#6D1305")
plt.scatter((V*1e-1)[0:5], F_N[0:5], color = "#C70039")
perr2 = np.sqrt(np.diag(pcov2))
plt.grid()
plt.xlabel(r"Volumen Sumergido [cm$^{3}$]")
plt.ylabel(r"Fuerza Normal [N]")
plt.text(14.5, 0.09, r"R$^{2}$ = 0.996", color = "#C70039", fontsize = 13)
plt.savefig("Normal.pdf")
m2, perr2[0], R2_Al2
```
