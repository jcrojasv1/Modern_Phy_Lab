```python
%matplotlib notebook
import numpy as np
import pandas as pd
import sympy as sym
from scipy import constants
import matplotlib.pyplot as plt
```

```python
f = np.array([200, 350, 400, 500, 650, 800, 
              900, 1000, 1200, 1400, 1600, 1800, 2000])
```

```python
v = np.array([3.28, 2.28, 2.35, 1.7, 1.34, 1.12, 1.09, 0.92,
             0.9, 0.8, 0.72, 0.66, 0.584])
```

```python
figFilter = plt.figure(figsize = (7, 5))
plt.scatter(f, v,label = "Muestreo", zorder = 2, color = "pink")
plt.axhline(max(v)/np.sqrt(2), linestyle = "dashed", color = "black", zorder = 1)
plt.axvline(350, linestyle = "dashed", color = "black", zorder = 1)
#plt.axvline(1100, linestyle = "dashed", color = "black", zorder = 1)
#plt.text(1250, max(v)/np.sqrt(2) + 0.07, r"$V_{pp-out} = \frac{mÃ¡x|V_{pp-out}|}{\sqrt{2}}$", fontsize = 15)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel(r"Amplitud del $V_{pp-out}$ [V]")
plt.title("Filtro Pasabajas RC")
plt.grid()
plt.legend()
```
