```python
%matplotlib notebook
import numpy as np
import pandas as pd
import sympy as sym
from scipy import constants
import matplotlib.pyplot as plt
```

```python
f = np.array([200, 350, 500, 650, 800, 1000,
              17000, 23000, 31000, 39000, 48000, 56000, 66000])
```

```python
v = np.array([0.536, 0.56, 0.576, 0.592, 0.608, 0.68,
             3.44, 3.84, 4.32, 4.56, 4.64, 4.8, 4.96])
```

```python
figFilter = plt.figure(figsize = (7, 5))
plt.scatter(f, v,label = "Muestreo", zorder = 2, color = "pink")
plt.axhline(max(v)/np.sqrt(2), linestyle = "dashed", color = "black", zorder = 1)
plt.axvline(17500, linestyle = "dashed", color = "black", zorder = 1)
#plt.axvline(1100, linestyle = "dashed", color = "black", zorder = 1)
plt.text(40000, max(v)/np.sqrt(2) + 0.35, r"$V_{pp-out} = \frac{mÃ¡x|V_{pp-out}|}{\sqrt{2}}$", fontsize = 15)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel(r"Amplitud del $V_{pp-out}$ [V]")
plt.title("Filtro Pasaaltas RL")
plt.grid()
plt.legend()
```
