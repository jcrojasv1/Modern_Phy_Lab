```python
%matplotlib notebook
import numpy as np
import pandas as pd
import sympy as sym
from scipy import constants
import matplotlib.pyplot as plt
```

```python
o = np.array([20, 30, 50, 70, 80, 90, 100, 200, 300, 400, 500, 600, 
              700, 800, 900, 1000, 1100, 2000])
```

```python
v = np.array([0.1, 0.16, 0.56, 0.68, 0.74, 0.82, 0.88, 1.12,
             1.18, 1.16, 1.12, 1.06, 1.02, 0.94, 0.9, 0.88,
             0.82, 0.52])
```

```python
figFilter = plt.figure(figsize = (7, 5))
plt.scatter(o, v,label = "Muestreo", zorder = 2, color = "pink")
plt.axhline(max(v)/np.sqrt(2), linestyle = "dashed", color = "black", zorder = 1)
plt.axvline(90, linestyle = "dashed", color = "black", zorder = 1)
plt.axvline(1100, linestyle = "dashed", color = "black", zorder = 1)
plt.text(1250, max(v)/np.sqrt(2) + 0.07, r"$V_{pp-out} = \frac{mÃ¡x|V_{pp-out}|}{\sqrt{2}}$", fontsize = 15)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel(r"Amplitud del $V_{pp-out}$ [V]")
plt.title("Filtro Pasabanda")
plt.grid()
plt.legend()
```
