```python
path = "C:/Users/Thomas/Physics_Labs/Electronics_for_Sciences/04 Informe Filtros/Datos/"
```

```python
def Change_of_time(data):
    index_list = np.where(data == min(data))
    for i in range(0, len(index_list[0]) - 1):
        index = i
        if i == 0:
            prev = index_list[0][0]
        else:
            if np.abs(prev - index_list[0][i]) < 20:
                prev = index_list[0][i]
            else:
                break
    mean = np.sum(index_list[0][0:index])/len(index_list[0][0:index])
    return int(mean)
```

# Análisis para frecuencia de 1 kHz:

Las próximas líneas de código muestran el análisis de las gráficas para el circuito RL bajo una frecuencia de $1$ kHz:

```python
frec_1kHz = pd.read_csv(path + "1 khz RL.csv")[["X1", "Y1", "X2", "Y2"]]
change_1kHz = frec_1kHz["X1"][2] - frec_1kHz["X1"][1]
frec_1kHz
```

```python
Vpp_1kHz = np.abs(max(frec_1kHz["Y1"]) - min(frec_1kHz["Y1"]))
print("El voltaje pico a pico de la fuente es: {} V".format(Vpp_1kHz))
```

```python
VCpp_1kHz = np.abs(max(frec_1kHz["Y2"]) - min(frec_1kHz["Y2"]))
print("El voltaje pico a pico del circuito es: {} V".format(VCpp_1kHz))
```

```python
dtC_1kHz = 80
dtF_1kHz = 110
time_1kHz = frec_1kHz["X2"][dtF_1kHz] - frec_1kHz["X2"][dtC_1kHz]
```

```python
fig1k = plt.figure(figsize = (7, 5))
plt.scatter(frec_1kHz["X1"][dtF_1kHz], min(frec_1kHz["Y1"]), color = "#D35400", zorder = 2)
plt.scatter(frec_1kHz["X2"][dtC_1kHz], min(frec_1kHz["Y2"]), color = "#1C4696", zorder = 2)
plt.plot(frec_1kHz["X2"], frec_1kHz["Y2"], linestyle = "dashed", label = "Circuito", zorder = 1)
plt.plot(frec_1kHz["X1"], frec_1kHz["Y1"], linestyle = "dashed", label = "Fuente de 2.5V", zorder = 1)
plt.xlabel("Tiempo en Segundos")
plt.ylabel("Amplitud de Voltaje")
plt.title("Gráfica para la Frecuencia de 1kHz")
plt.grid()
plt.legend()
```

```python
print("El desfase temporal es de cerca de {} segundos.".format(time_1kHz))
```

```python
phase_1kHz = time_1kHz*2*np.pi*1000 * (180/(np.pi))
print("El desfase angular es de aproximadamente {} grados.".format(phase_1kHz))
```
