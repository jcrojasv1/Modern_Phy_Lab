```python
from openpyxl import*
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Hoja 0.457mm
book= load_workbook('datosoficiales2.xlsx')
Hoja=book.active

```

```python

def funcion(filename,longitud):
    book= load_workbook(filename)
    Hoja=book.active
    j=1
    lista_x_doble=[]
    for i in range(3,longitud):
            distancia=float(Hoja.cell(i,j).value)
            m1=float(Hoja.cell(i,j+1).value)
            m2=float(Hoja.cell(i,j+2).value)
            m3=float(Hoja.cell(i,j+3).value)
            m4=float(Hoja.cell(i,j+4).value)
            m5=float(Hoja.cell(i,j+5).value)
            m6=float(Hoja.cell(i,j+6).value)
            promedio=(m1+m2+m3+m4+m5+m6)/6
            valor=[distancia,promedio]
            lista_x_doble.append(valor)

    lista_x_doble = np.array(lista_x_doble)
    lista_x_izquierda=[]
    k=9
    for i in range(3,longitud):
            distancia=float(Hoja.cell(i,k).value)
            m1=float(Hoja.cell(i,k+1).value)
            m2=float(Hoja.cell(i,k+2).value)
            m3=float(Hoja.cell(i,k+3).value)
            m4=float(Hoja.cell(i,k+4).value)
            m5=float(Hoja.cell(i,k+5).value)
            m6=float(Hoja.cell(i,k+6).value)
            promedio=(m1+m2+m3+m4+m5+m6)/6
            valor=[distancia,promedio]
            lista_x_izquierda.append(valor)
    lista_x_izquierda = np.array(lista_x_izquierda)
    lista_x_derecha=[]
    h=17
    for i in range(3,longitud):
            distancia=float(Hoja.cell(i,h).value)
            m1=float(Hoja.cell(i,h+1).value)
            m2=float(Hoja.cell(i,h+2).value)
            m3=float(Hoja.cell(i,h+3).value)
            m4=float(Hoja.cell(i,h+4).value)
            m5=float(Hoja.cell(i,h+5).value)
            m6=float(Hoja.cell(i,h+6).value)
            promedio=(m1+m2+m3+m4+m5+m6)/6
            valor=[distancia,promedio]
            lista_x_derecha.append(valor)        
    lista_x_derecha = np.array(lista_x_derecha)
    book.close()
    return lista_x_doble,lista_x_izquierda,lista_x_derecha
```

```python
lista_x_doble,lista_x_izquierda,lista_x_derecha= funcion('datosoficiales2.xlsx',304)

plt.figure(figsize=(12,8))
plt.plot(lista_x_doble[:,0]-3,lista_x_doble[:,1],label="Datos para doble rendija")
plt.plot(lista_x_izquierda[:,0]-3,lista_x_izquierda[:,1],label="Datos para rendija izquierda")
plt.plot(lista_x_derecha[:,0]-3,lista_x_derecha[:,1],label="Datos para rendija derecha")
plt.grid()
plt.legend()
plt.title("Patrón de interferencia para $d = 0.457 \: mm$")
plt.xlabel("Posición x [$mm$]")
plt.ylabel("Conteos")
plt.savefig("fig1.png")
```

```python
from scipy.optimize import curve_fit
d1=0.457
```

```python
def intensidad_1(theta,i_0,lamb):
    i = i_0*(np.cos((np.pi/lamb)*d1*theta))**2
    return i
```

```python
def f_determinar(x,i_0,lamb):
    d=0.457
    D1=500
    D2=500
    y_min=-3
    y_max=3
    dy=y_max-y_min/10000
    y=np.linspace(y_min,y_max,10000)
    f = []
    for x_i in x:
        k=(np.exp((1j*np.pi*(-y)**2)/D1*lamb))*(np.exp((1j*np.pi*(x_i-y)**2)/D2*lamb))
        multiplo=np.exp((2*np.pi*1j*(D1+D2)/lamb))
        integral=np.trapz(k,y,dy)
        fnc=(abs(multiplo*integral))
        f.append(fnc)
    fnc = i_0*np.array(f)
    return fnc          
```

```python
popt1, pcov1 = curve_fit(f_determinar,lista_x_izquierda[:,0]-3,lista_x_izquierda[:,1])
```

```python
y_opt = f_determinar(lista_x_izquierda[:,0]-3,popt1[0],popt1[1])
```
