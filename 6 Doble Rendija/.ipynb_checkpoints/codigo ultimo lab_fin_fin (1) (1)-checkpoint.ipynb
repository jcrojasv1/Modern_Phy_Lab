```python
from openpyxl import*
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
book= load_workbook('datosoficiales2.xlsx')
Hoja=book.active
```

```python
def funcion(filename,longitud):
    book= load_workbook(filename)
    Hoja=book.active
    j=1
    lista_x_doble=[]
    for i in range(3,longitud):
            distancia=float(Hoja.cell(i,j).value)
            m1=float(Hoja.cell(i,j+1).value)
            m2=float(Hoja.cell(i,j+2).value)
            m3=float(Hoja.cell(i,j+3).value)
            m4=float(Hoja.cell(i,j+4).value)
            m5=float(Hoja.cell(i,j+5).value)
            m6=float(Hoja.cell(i,j+6).value)
            promedio=(m1+m2+m3+m4+m5+m6)/6
            valor=[distancia,promedio]
            lista_x_doble.append(valor)

    lista_x_doble = np.array(lista_x_doble)
    lista_x_izquierda=[]
    k=9
    for i in range(3,longitud):
            distancia=float(Hoja.cell(i,k).value)
            m1=float(Hoja.cell(i,k+1).value)
            m2=float(Hoja.cell(i,k+2).value)
            m3=float(Hoja.cell(i,k+3).value)
            m4=float(Hoja.cell(i,k+4).value)
            m5=float(Hoja.cell(i,k+5).value)
            m6=float(Hoja.cell(i,k+6).value)
            promedio=(m1+m2+m3+m4+m5+m6)/6
            valor=[distancia,promedio]
            lista_x_izquierda.append(valor)
    lista_x_izquierda = np.array(lista_x_izquierda)
    lista_x_derecha=[]
    h=17
    for i in range(3,longitud):
            distancia=float(Hoja.cell(i,h).value)
            m1=float(Hoja.cell(i,h+1).value)
            m2=float(Hoja.cell(i,h+2).value)
            m3=float(Hoja.cell(i,h+3).value)
            m4=float(Hoja.cell(i,h+4).value)
            m5=float(Hoja.cell(i,h+5).value)
            m6=float(Hoja.cell(i,h+6).value)
            promedio=(m1+m2+m3+m4+m5+m6)/6
            valor=[distancia,promedio]
            lista_x_derecha.append(valor)        
    lista_x_derecha = np.array(lista_x_derecha)
    book.close()
    return lista_x_doble,lista_x_izquierda,lista_x_derecha
```

```python
lista_x_doble,lista_x_izquierda,lista_x_derecha= funcion('datosoficiales2.xlsx',304)

plt.figure(figsize=(12,8))
plt.plot(lista_x_doble[:,0]-3,lista_x_doble[:,1],label="Datos para doble rendija")
plt.plot(lista_x_izquierda[:,0]-3,lista_x_izquierda[:,1],label="Datos para rendija izquierda")
plt.plot(lista_x_derecha[:,0]-3,lista_x_derecha[:,1],label="Datos para rendija derecha")
plt.grid()
plt.legend()
plt.title("Patrón de interferencia para $d = 0.457 \: mm$")
plt.xlabel("Posición x [$mm$]")
plt.ylabel("Conteos")
plt.savefig("fig1.png")
```

```python
L=500 ## mm

def f(x):
    fnc= np.arctan2(x,L)
    return fnc
def maxi(x):
    maximo = 0
    valor = 0
    for i in range(len(x)):
        if x[i][1] > maximo:
            maximo = x[i][1]
            valor = x[i][0]
    return valor
#valores doble rendija
valoresdoble=f(lista_x_doble[:,0]-maxi(lista_x_doble))
valoresdoble_1=f(lista_x_doble1[:,0]-maxi(lista_x_doble1))
valoresdoble_2=f(lista_x_doble2[:,0]-maxi(lista_x_doble2))
#valores rendija izquierda
valoresiz=f(lista_x_izquierda[:,0]-maxi(lista_x_izquierda))
valoresiz_1=f(lista_x_izquierda1[:,0]-maxi(lista_x_izquierda1))
valoresiz_2=f(lista_x_izquierda2[:,0]-maxi(lista_x_izquierda2))

#valores rendija derecha
valoresde=f(lista_x_derecha[:,0]-maxi(lista_x_derecha))
valoresde_1=f(lista_x_derecha1[:,0]-maxi(lista_x_derecha1))
valoresde_2=f(lista_x_derecha2[:,0]-maxi(lista_x_derecha2))
```

```python
def incertidumbre_theta(x, l, delta_x):
    delta_t = np.abs(1/(l*(1+(x/l)**2)))*delta_x
    return delta_t
```

```python
delta_y = 1 # eventos
delta_x = 0.05 ## mm
```

```python
delta_theta_doble = incertidumbre_theta(valoresdoble, L, delta_x)
delta_theta_izq = incertidumbre_theta(valoresiz, L, delta_x)
delta_theta_der = incertidumbre_theta(valoresde, L, delta_x)

plt.figure(figsize=(12,8))
plt.errorbar(valoresdoble, lista_x_doble[:,1], yerr=delta_y, xerr=delta_theta_doble, fmt=".", label="Datos (doble rendija)")
plt.errorbar(valoresiz, lista_x_izquierda[:,1], yerr=delta_y, xerr=delta_theta_izq, fmt=".", label="Datos (rendija izquierda)")
plt.errorbar(valoresde, lista_x_derecha[:,1], yerr=delta_y, xerr=delta_theta_der, fmt=".", label="Datos (rendija derecha)")
plt.grid()
plt.legend()
plt.title("Patrón de interferencia para $d = 0.457 \: mm$")
plt.xlabel("Posición " + r"$\theta$" + " [rad]")
plt.ylabel("Conteos")
plt.savefig("fig1_rad.png")
```

```python
d1 = 0.457 ## mm
d2 = 0.406 ## mm
d3 = 0.356 ## mm
a = 0.1 # mm
```

```python
lamb=0.000546
def intensidad_1(theta,i_0):
    i = i_0*(np.cos((np.pi/(lamb))*d1*np.sin(theta)))**2
    return i

def intensidad_2(theta,i_0):
    multiplo = i_0*(np.cos((np.pi/(lamb))*d1*np.sin(theta)))**2
    valor=multiplo*i_0*((np.sin(((np.pi*a)/lamb)*np.sin(theta))/(((np.pi*a)/lamb)*np.sin(theta))))**2
    
    return valor
```

```python
popt1, pcov1 = curve_fit(intensidad_2,valoresdoble,lista_x_doble[:,1])
```

```python
y_new1 = intensidad_2(valoresdoble,20*popt1)
```
